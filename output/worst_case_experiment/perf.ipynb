```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
normal_df = pd.read_csv("splay_tree.csv")

# Plotting the line graph for normal splay tree performance
plt.figure(figsize=(20, 6))
plt.plot(normal_df['query_time'], label='Normal Splay Tree', linestyle='-', marker='o')

# Add titles and labels
plt.title('Query Time Performance of Normal Splay Tree', fontsize=14)
plt.xlabel('Query Index', fontsize=12)
plt.ylabel('Query Time (ms)', fontsize=12)
plt.ylim(top=20)

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.legend()
plt.show()
```
