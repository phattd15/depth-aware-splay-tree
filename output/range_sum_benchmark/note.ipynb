```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Import numpy to calculate log2

# Load the CSV file into a DataFrame
filename = "results_log.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
pbds_times = data["std::set"]  # Assuming the PBDS times column is named "PBDS"
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Calculate log2 of tree sizes
log_tree_sizes = np.log2(tree_sizes)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot each line
tree_sizes_sorted_indices = log_tree_sizes.argsort()  # Get sorted indices
sorted_log_tree_sizes = log_tree_sizes.iloc[tree_sizes_sorted_indices]
sorted_pbds_times = pbds_times.iloc[tree_sizes_sorted_indices]
sorted_depth_aware_splay_tree_times = depth_aware_splay_tree_times.iloc[tree_sizes_sorted_indices]

ax.plot(sorted_log_tree_sizes, sorted_pbds_times, label="std::set", color="steelblue", marker='o')
ax.plot(sorted_log_tree_sizes, sorted_depth_aware_splay_tree_times, label="Depth-Aware Splay Tree", color="green", marker='o')

# Add labels, title, and legend
ax.set_xlabel("log2(Tree Size)")
ax.set_ylabel("Average Query Time (μs)")
ax.set_title("Benchmarking Results: Range Sum Query by log2(Tree Size)")
ax.legend()
ax.set_ylim(bottom=0)

# Add grid lines for better readability
ax.grid(axis="both", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Import numpy to calculate log2

# Load the CSV file into a DataFrame
filename = "results_log.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
pbds_times = data["PBDS"]  # Assuming the PBDS times column is named "PBDS"
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Calculate log2 of tree sizes
log_tree_sizes = np.log2(tree_sizes)

# Calculate the ratio of DAST to PBDS in percentage
time_ratio = 100 * depth_aware_splay_tree_times / pbds_times

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the ratio as a line chart
ax.plot(log_tree_sizes, time_ratio, label="DAST / PBDS (%)", color="purple", marker='o')

# Add labels, title, and legend
ax.set_xlabel("log2(Tree Size)")
ax.set_ylabel("Time Ratio (DAST / PBDS) in %")
ax.set_title("Percentage Ratio of Depth-Aware Splay Tree to PBDS Order Of Key Query Query Times by Log2(Tree Size)")
ax.legend()

# Set Y-axis minimum to 0
ax.set_ylim(0, 100)

# Add grid lines for better readability
ax.grid(axis="both", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
