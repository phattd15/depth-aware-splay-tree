```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a DataFrame
filename = "results.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
std_set_times = data["std::set"]
original_splay_tree_times = data["OriginalSplayTree"]
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Prepare data for grouped bar plot
labels = tree_sizes.astype(str)  # Convert tree sizes to strings for x-axis labels
x = np.arange(len(labels))  # Label locations
width = 0.25  # Width of bars

# Create the plot
fig, ax = plt.subplots(figsize=(8, 3))

# Plot each group
ax.bar(x - width, std_set_times, width, label="std::set", color="steelblue")
ax.bar(x, original_splay_tree_times, width, label="Original Splay Tree", color="orange")
ax.bar(x + width, depth_aware_splay_tree_times, width, label="Depth-Aware Splay Tree", color="green")

# Add labels, title, and legend
ax.set_xlabel("Tree Size")
ax.set_ylabel("Average Query Time (μs)")
ax.set_title("Benchmarking Results: Query Times by Random Tree Size")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Add grid lines for better readability
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
