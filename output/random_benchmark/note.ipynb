```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a DataFrame
filename = "results.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
std_set_times = data["std::set"]
original_splay_tree_times = data["OriginalSplayTree"]
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Prepare data for grouped bar plot
labels = tree_sizes.astype(str)  # Convert tree sizes to strings for x-axis labels
x = np.arange(len(labels))  # Label locations
width = 0.25  # Width of bars

# Create the plot
fig, ax = plt.subplots(figsize=(8, 3))

# Plot each group
ax.bar(x - width, std_set_times, width, label="std::set", color="steelblue")
ax.bar(x, original_splay_tree_times, width, label="Original Splay Tree", color="orange")
ax.bar(x + width, depth_aware_splay_tree_times, width, label="Depth-Aware Splay Tree", color="green")

# Add labels, title, and legend
ax.set_xlabel("Tree Size")
ax.set_ylabel("Average Query Time (s)")
ax.set_title("Benchmarking Results: Query Times by Random Tree Size")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Add grid lines for better readability
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
filename = "results_small.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
std_set_times = data["std::set"]
original_splay_tree_times = data["OriginalSplayTree"]
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot each line
tree_sizes_sorted = tree_sizes.sort_values()
ax.plot(tree_sizes_sorted, std_set_times, label="std::set", color="steelblue", marker='o')
ax.plot(tree_sizes_sorted, original_splay_tree_times, label="Original Splay Tree", color="orange", marker='o')
ax.plot(tree_sizes_sorted, depth_aware_splay_tree_times, label="Depth-Aware Splay Tree", color="green", marker='o')

# Add labels, title, and legend
ax.set_xlabel("Tree Size")
ax.set_ylabel("Average Query Time (s)")
ax.set_title("Benchmarking Results: Query Times by Random Tree Size")
ax.legend()
ax.set_ylim(bottom=0)

# Add grid lines for better readability
ax.grid(axis="both", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
filename = "results_small.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
std_set_times = data["std::set"]
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Calculate the ratio of std::set to Depth-Aware Splay Tree
time_ratio = 100 * depth_aware_splay_tree_times / std_set_times

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the ratio as a line chart
tree_sizes_sorted = tree_sizes.sort_values()
ax.plot(tree_sizes_sorted, time_ratio, label="std::set / Depth-Aware Splay Tree", color="purple", marker='o')

# Add labels, title, and legend
ax.set_xlabel("Tree Size")
ax.set_ylabel("Time Ratio (DAST / std::set) in percentage")
ax.set_title("Ratio of  Depth-Aware Splay Tree Query Times to std::set by Tree Size")
ax.legend()

# Set Y-axis minimum to 0
ax.set_ylim(bottom=0)

# Add grid lines for better readability
ax.grid(axis="both", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Import numpy to calculate log2

# Load the CSV file into a DataFrame
filename = "results_log.csv"
data = pd.read_csv(filename)

# Extract columns for plotting
tree_sizes = data["TreeSize"].astype(int)
std_set_times = data["std::set"]
original_splay_tree_times = data["OriginalSplayTree"]
depth_aware_splay_tree_times = data["DepthAwareSplayTree"]

# Calculate log2 of tree sizes
log_tree_sizes = np.log2(tree_sizes)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot each line
tree_sizes_sorted = log_tree_sizes.sort_values()
ax.plot(tree_sizes_sorted, std_set_times, label="std::set", color="steelblue", marker='o')
ax.plot(tree_sizes_sorted, original_splay_tree_times, label="Original Splay Tree", color="orange", marker='o')
ax.plot(tree_sizes_sorted, depth_aware_splay_tree_times, label="Depth-Aware Splay Tree", color="green", marker='o')

# Add labels, title, and legend
ax.set_xlabel("log2(Tree Size)")
ax.set_ylabel("Average Query Time (s)")
ax.set_title("Benchmarking Results: Query Times by log2(Tree Size)")
ax.legend()
ax.set_ylim(bottom=0)

# Add grid lines for better readability
ax.grid(axis="both", linestyle="--", alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
