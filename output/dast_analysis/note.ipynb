```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV data
df = pd.read_csv("dast_data.csv")

# Set up the plot area
plt.figure(figsize=(20, 18))

# Plot 1: Depth Threshold vs Average Query Time (microseconds)
plt.subplot(3, 1, 1)  # (nrows, ncols, index)
plt.plot(df['depth_threshold'], df['avg_query_time(microseconds)'], label='Avg Query Time', linestyle='-', marker='o', color='b')
plt.title('Depth Threshold vs Avg Query Time', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Avg Query Time (microseconds)', fontsize=12)
plt.grid(True)
plt.legend()

# Plot 2: Depth Threshold vs Avg Depth
plt.subplot(3, 1, 2)
plt.plot(df['depth_threshold'], df['avg_depth'], label='Avg Depth', linestyle='-', marker='o', color='g')
plt.title('Depth Threshold vs Avg Depth', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Avg Depth', fontsize=12)
plt.grid(True)
plt.legend()

# Plot 3: Depth Threshold vs Splay Count
plt.subplot(3, 1, 3)
plt.plot(df['depth_threshold'], df['splay_count'], label='Splay Count', linestyle='-', marker='o', color='r')
plt.title('Depth Threshold vs Splay Count', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Splay Count', fontsize=12)
plt.grid(True)
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Assuming df is already defined and contains the relevant data
plt.figure(figsize=(12, 6))

# Plot the main line
plt.plot(df['depth_threshold'], df['avg_query_time(microseconds)'], label='Avg Query Time', linestyle='-', marker='o')

# Highlight the point where threshold=27
threshold_value = 27
query_time_at_threshold = df[df['depth_threshold'] == threshold_value]['avg_query_time(microseconds)'].iloc[0]
plt.scatter(threshold_value, query_time_at_threshold, color='red', label=f'Threshold = {threshold_value}', zorder=5)

# Add title, labels, grid, and legend
plt.title('Depth Threshold vs Avg Query Time', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Avg Query Time (microseconds)', fontsize=12)
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df['depth_threshold'], df['avg_depth'], label='Avg Depth', linestyle='-', marker='o')
plt.title('Depth Threshold vs Avg Depth', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Avg Depth', fontsize=12)
plt.grid(True)

# Highlight the point where threshold=27
threshold_value = 27
pp = df[df['depth_threshold'] == threshold_value]['avg_depth'].iloc[0]
plt.scatter(threshold_value, pp, color='red', label=f'Threshold = {threshold_value}', zorder=5)

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df['depth_threshold'], df['splay_count'], label='Splay Count', linestyle='-', marker='o')
plt.title('Depth Threshold vs Splay Count', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Splay Count', fontsize=12)
plt.grid(True)

threshold_value = 27
pp = df[df['depth_threshold'] == threshold_value]['splay_count'].iloc[0]
plt.scatter(threshold_value, pp, color='red', label=f'Threshold = {threshold_value}', zorder=5)

plt.legend()

plt.show()
```

```python
splay_count_log2 = np.where(df['splay_count'] == 0, 0, np.log2(df['splay_count']))
plt.figure(figsize=(12, 6))

plt.plot(df['depth_threshold'], splay_count_log2, label='Splay Count', linestyle='-', marker='o')
plt.title('Depth Threshold vs Log2(Splay Count)', fontsize=14)
plt.xlabel('Depth Threshold', fontsize=12)
plt.ylabel('Log2(Splay Count)', fontsize=12)

threshold_value = 27
pp = df[df['depth_threshold'] == threshold_value]['splay_count'].iloc[0]
plt.scatter(threshold_value, np.log2(pp), color='red', label=f'Threshold = {threshold_value}', zorder=5)
plt.grid(True)
plt.legend()

plt.show()
```
